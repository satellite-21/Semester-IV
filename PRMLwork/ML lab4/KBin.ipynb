```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_table("C:\\Users\\Kartik\\Desktop\\dataset.txt")
```

```python
df
```

```python
df1 = df.copy()
```

```python
bins = 4
#width would be (max-min)/bins
```

```python
col = list(df1.columns)
```

```python
df1
```

```python
#FOR AREA
lower_bound = int(min(df['Area']))
m = min(df['Area'])
M = max(df['Area'])
width = (M-m)/bins
for i in range(4):
    df1.loc[(df1['Area'] >= lower_bound + i*width) & (df1['Area'] < lower_bound + (i+1)*width), 'Area'] = i
```

```python
df1
```

```python
#FOR PERIMETER
lower_bound = int(min(df['Perimeter']))
m = min(df['Perimeter'])
M = max(df['Perimeter'])
width = (M-m)/bins
for i in range(4):
    df1.loc[(df1['Perimeter'] >= lower_bound + i*width) & (df1['Perimeter'] < lower_bound + (i+1)*width), 'Perimeter'] = i
df1
```
